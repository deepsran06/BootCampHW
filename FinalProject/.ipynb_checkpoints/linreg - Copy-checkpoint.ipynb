```python
# Import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load in CSVs
HOF = pd.read_csv('hall_of_fame.csv')
NBA = pd.read_csv('not_hof.csv')
all_players = pd.read_csv('all_players_data.csv')
```

```python
# Fill NaN values with 0
all_players = all_players.fillna(0)
all_players
```

```python
# Make a copy of the dataframe
copy = all_players
copy
```

```python
# Plot the datasets in a boxplot
fig = plt.figure(figsize=(16,16))
for counter, column in enumerate(['TRB', 'AST', 'STL', 'BLK', 'PTS','VORP','WS','BPM',
       'all_star_appear', 'nba_champ', 'all_nba']):
    plt.subplot(4,4,counter+1)
    plt.subplots_adjust(wspace=0.5,hspace=0.3)
    ax = sns.boxplot(x="HOF", y=column, data=all_players)

plt.show()
```

```python
# Assign the data to X and y

X = all_players[['WS', 'BPM', 'VORP', '3P', '2P', 'TRB',
         'AST', 'BLK', 'TOV', 'PTS', 'all_star_appear',
         'nba_champ', 'all_defensive', 'all_star_mvp',
         'league_mvp', 'scoring_champ', 'all_nba', 'finals_mvp',
         'def_poy', 'stl_champ', 'ast_champ', 'trb_champ',
         'blk_champ']]
y = all_players['HOF'].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
### BEGIN SOLUTION
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

### END SOLUTION 

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
X_NBA = all_players[['WS', 'BPM', 'VORP', '3P', '2P', 'TRB',
         'AST', 'BLK', 'TOV', 'PTS', 'all_star_appear',
         'nba_champ', 'all_defensive', 'all_star_mvp',
         'league_mvp', 'scoring_champ', 'all_nba', 'finals_mvp',
         'def_poy', 'stl_champ', 'ast_champ', 'trb_champ',
         'blk_champ']]
y_NBA = all_players['HOF']
```
